```python
import pandas as pd
```

```python
PATH = '../data'
```

#### Customer_Master_Data

```python
cust_df = pd.read_csv(f'{PATH}/customer_master_data.csv')
cust_df.head()
```

```python
column_to_drop = cust_df.columns[0]
cust_df = cust_df.drop(columns=column_to_drop)
```

```python
cust_df.to_csv(f'{PATH}/customer_master_data.csv', index=False)
```

```python
cust_df.head()
```

```python
print(f"Shape: {cust_df.shape}")
cust_df.isnull().sum()
```
